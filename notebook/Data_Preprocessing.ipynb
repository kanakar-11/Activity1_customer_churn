```python
import pandas as pd
import numpy as np # Imports the NumPy library and assigns it to the alias 'np'

# Load the dataset
df = pd.read_csv('C:/Users/kanak/OneDrive/Desktop/ML ops/Activity1_customer_churn/data/churn.csv')
df 
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handle missing values (if any)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
# Calculate the mean only for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
# Convert categorical variables to dummy variables
df = pd.get_dummies(df, drop_first=True)
```

```python
# Save the preprocessed data
df.to_csv('C:/Users/kanak/Downloads/churn.csv', index=False)
```
